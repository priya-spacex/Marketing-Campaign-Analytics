```python
# Cell 2 — Complete Initialization, Data Load & Preprocessing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust pandas display rules to ensure all columns remain visible during profiling
pd.set_option('display.max_columns', None)
sns.set_style("darkgrid")

# Load the newly uploaded file into our dataframe
df = pd.read_csv('45.csv')

# Clean up our string data, convert timelines, and engineer new marketing metrics
df['Acquisition_Cost'] = df['Acquisition_Cost'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df['Acquisition_Cost'] = pd.to_numeric(df['Acquisition_Cost'], errors='coerce')

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Engagement Rate'] = (df['Clicks'] / df['Impressions']) * 100
df['Total Conversions'] = df['Clicks'] * df['Conversion_Rate']

df['Engagement Rate'] = df['Engagement Rate'].fillna(0)
df['Total Conversions'] = df['Total Conversions'].fillna(0)

print(f"Data Successfully Synchronized! Shape: {df.shape[0]} rows, {df.shape[1]} columns")
df[['Date', 'Acquisition_Cost', 'Engagement Rate', 'Total Conversions']].head()
```

```python
# Cell 3 — High-Level Marketing KPI Extraction

# Calculate financial and performance aggregations across the entire dataset
avg_conversion = df['Conversion_Rate'].mean()
avg_engagement = df['Engagement Rate'].mean()
total_spend = df['Acquisition_Cost'].sum()
total_conversions_delivered = df['Total Conversions'].sum()

# Print out a clean executive summary dashboard
print("==================================================")
print("       EXECUTIVE MARKETING KPI DASHBOARD          ")
print("==================================================")
print(f"Average Campaign Conversion Rate : {avg_conversion * 100:.2f}%")
print(f"Average Audience Engagement Rate : {avg_engagement:.2f}%")
print(f"Total Portfolio Marketing Spend  : ${total_spend:,.2f}")
print(f"Total Acquired Customers Volume  : {int(total_conversions_delivered):,}")
print("==================================================")
```

```python
# Cell 4 — Strategic Campaign Type Analysis

# Group data by campaign type and compute performance means, sorted by profitability
campaign_analysis = df.groupby('Campaign_Type')[[
    'ROI',
    'Conversion_Rate',
    'Engagement Rate'
]].mean().sort_values(by='ROI', ascending=False)

# Display the strategic data summary table
campaign_analysis
```

```python
# Cell 5 — Static Campaign Profitability Bar Chart

plt.figure(figsize=(10, 6))

# Generate a bar plot using Seaborn's library
sns.barplot(
    x=campaign_analysis.index,
    y=campaign_analysis['ROI'],
    palette='viridis' # This applies a nice modern color spectrum across our bars
)

# Label our chart clearly for stakeholders
plt.title("Average ROI Performance by Campaign Strategy", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Campaign Strategic Type", fontsize=12, labelpad=10)
plt.ylabel("Average Financial ROI Multiplier", fontsize=12, labelpad=10)

# Cleanly adjust our visual layout boundaries before drawing the plot
plt.tight_layout()
plt.show()
```

```python
# Cell 6 — Marketing Channel Performance Analysis

# Group your records by media network, computing average ROI and Conversion Rates
channel_analysis = df.groupby('Channel_Used')[['ROI', 'Conversion_Rate']].mean().sort_values(by='ROI', ascending=False)

# Render the channel insights summary table
channel_analysis
```

```python
# Cell 7 — Marketing Channel ROI Comparison Plot

plt.figure(figsize=(10, 6))

# Construct a distinct statistical bar chart for our channels
sns.barplot(
    x=channel_analysis.index,
    y=channel_analysis['ROI'],
    palette='magma' # Changing the palette helps separate different visual findings
)

# Customize the typography and grid layout properties
plt.title("Average ROI Analysis across Marketing Channels", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Digital Distribution Channel", fontsize=12, labelpad=10)
plt.ylabel("Average ROI Multiplier", fontsize=12, labelpad=10)

plt.tight_layout()
plt.show()
```

```python
# Cell 8 — Advanced Financial Metric Engineering

# 1. Safely compute Cost Per Click (CPC), avoiding division by zero
df['Cost Per Click (CPC)'] = np.where(df['Clicks'] > 0, df['Acquisition_Cost'] / df['Clicks'], 0)

# 2. Safely compute Cost Per Acquisition (CPA), avoiding division by zero
df['Cost Per Acquisition'] = np.where(df['Total Conversions'] > 0, df['Acquisition_Cost'] / df['Total Conversions'], 0)

print("Advanced Feature Engineering Complete.")
# Review our newly created financial tracking metrics
df[['Acquisition_Cost', 'Clicks', 'Cost Per Click (CPC)', 'Total Conversions', 'Cost Per Acquisition']].head()
```
